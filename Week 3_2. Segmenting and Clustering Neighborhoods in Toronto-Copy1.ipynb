```python
import pandas as pd

url = 'https://raw.githubusercontent.com/FernandoDelgadoChaves/Coursera_Capstone/main/df1.csv'
df1 = pd.read_csv(url, error_bad_lines=False)
df1.columns = ['Postal Code', 'Borough', 'Neighborhood']
df1
```

In this case, I extracted csv with Toronto geographical coordinates to dataframe.

```python
toronto_geocsv = 'https://cocl.us/Geospatial_data'
!wget -q -O 'toronto_m.geospatial_data.csv' toronto_geocsv
geocsv_data = pd.read_csv(toronto_geocsv)
geocsv_data.head()
```

```python
df = pd.concat([df1, geocsv_data],axis=1)
df.head()

df.columns = ['Postal Code','Borough','Neighbourhood', 'Postcode', 'Latitude', 'Longitude']
df.drop(columns=['Postcode'], inplace= True)
df.head(10)
```
